```python
import re
import pandas as pd
df = pd.read_csv("texts.csv")
```

```python
df
```

```python
for i, row in df.iterrows():
    text = row["text"]

    variants = re.findall(r"rs\d+", text)
    genes = re.findall(r"\b[A-Z0-9]{3,6}\b", text)
    diseases = re.findall(r"[A-Z][a-zA-Z’']+(?: [A-Z][a-zA-Z’']+)*", text)

    print(row["id"], variants, genes, diseases)
```

we can see that the regex extraction workds ok for extracting variants, but doesn't work so well for gene and phenotype extraction, hence we need to use a more semantic approach to efficiently extractt these information
